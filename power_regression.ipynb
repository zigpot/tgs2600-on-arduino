```python
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 0.95, 0.9, 0.8, 0.75, .7] #Rs/Ro
y = [3, 5, 10, 30, 40, 80] #ppm
popt, pcov = curve_fit(lambda fx,a,b: a*fx**-b,  x,  y)
power_y = popt[0]*x**-popt[1]

plt.xlabel('rs/ro')
plt.ylabel('concentration (ppm)')
plt.scatter(x, y, label='data points', c= 'red', alpha = 0.5, s = 18)
plt.plot(x, power_y, label='power fit regression')
plt.legend()
plt.show()
```

```python
popt, pcov = curve_fit(lambda fx,a,b: a*fx**-b,  x,  y)
x_linspace = np.linspace(min(x), max(x), 100)
power_y = popt[0]*x_linspace**-popt[1]

plt.xlabel('rs/ro')
plt.ylabel('concentration (ppm)')
plt.scatter(x, y, label='data points', c= 'red', alpha = 0.5, s = 18)
plt.plot(x_linspace, power_y, label='power fit regression (smooth)')
plt.legend()
plt.show()
```

```python
print('The regressions equation is {:.2f}x^-{:.2f}'
#print('The regressions equation is {}x^-{}'
.format(popt[0],popt[1]))
```

```python
plt.xlabel('rs/ro')
plt.ylabel('concentration (ppm)')
plt.scatter(x, y, label='data points', c= 'red', alpha = 0.5, s = 18)
plt.loglog(x_linspace, power_y, label='power fit regression (smooth)')
plt.legend()
plt.show()
```
